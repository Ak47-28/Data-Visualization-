```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("iris.csv")
```

```python
ax=plt.subplots(1,1,figsize=(10,8))
sns.countplot('Species',data=iris)
plt.title("Iris Species Count")
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
ax=plt.subplots(1,1,figsize=(10,8))
iris['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Iris Species %")
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(x = 'SepalLengthCm', y = 'SepalWidthCm', data = iris)
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue ="Species",
              height = 6).map(plt.scatter,
                              'SepalWidthCm',
                              'PetalLengthCm').add_legend()
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
### code here
plt.style.use('ggplot')
plt.subplot(2,2,1)
sns.boxplot(x = 'Species', y = 'SepalLengthCm', data = iris)
plt.subplot(2,2,2)
sns.boxplot(x = 'Species', y = 'SepalWidthCm', data = iris)
plt.subplot(2,2,3)
sns.boxplot(x = 'Species', y = 'PetalLengthCm', data = iris)
plt.subplot(2,2,4)
sns.boxplot(x = 'Species', y = 'PetalWidthCm', data = iris)
```

###### 6. Create a Strip plot 

```python
### code here
plt.subplot(2,2,1)
sns.stripplot(x = 'Species', y = 'SepalLengthCm', data = iris, jitter = True)
plt.subplot(2,2,2)
sns.stripplot(x = 'Species', y = 'SepalWidthCm', data = iris, jitter = True)
plt.subplot(2,2,3)
sns.stripplot(x = 'Species', y = 'PetalLengthCm', data = iris, jitter = True)
plt.subplot(2,2,4)
sns.stripplot(x = 'Species', y = 'PetalWidthCm', data = iris, jitter = True)
```

###### 7. Create a Violin Plot.

```python
### code here
plt.style.use('ggplot')
plt.subplot(2,2,1)
sns.violinplot(x = 'Species', y = 'SepalLengthCm', data = iris)
plt.subplot(2,2,2)
sns.violinplot(x = 'Species', y = 'SepalWidthCm', data = iris)
plt.subplot(2,2,3)
sns.violinplot(x = 'Species', y = 'PetalLengthCm', data = iris)
plt.subplot(2,2,4)
sns.violinplot(x = 'Species', y = 'PetalWidthCm', data = iris)
```
